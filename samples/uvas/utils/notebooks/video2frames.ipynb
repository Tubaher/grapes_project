```
import cv2
```

```
FPS = 60 # frames per second of the target video
IMG_STEP = 5 #Interval from image to image (in secs) 
FILE = 'Hilera_2_2_36_2.MOV' #name of the video
OUTPUT_DIR = 'frames/'
```

```
vidcap = cv2.VideoCapture(FILE)
success,image = vidcap.read()
count = 0
while success:
  success,image = vidcap.read()
  if count%(FPS*IMG_STEP) == 0:
    cv2.imwrite(OUTPUT_DIR + 'frame%d.jpg' % count, image)     # save frame as JPEG file      
    print('Saved new frame', count)
  count += 1
```
